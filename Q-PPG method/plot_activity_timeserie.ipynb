```python
sns.set(rc={'figure.figsize':(28,8)})
sns.set(rc={'axes.facecolor':'white', 'figure.facecolor':'white'})
sns.set(rc={'font.family' : 'Times New Roman'})
```

```python
sns.set(font_scale = 1.8)
with sns.axes_style("ticks"):
  r = sns.lineplot(x=np.arange(0,len(activity[:,0])), y=activity[:,0], linewidth=1, legend=False)
  r.set_xlabel("Temps (en fenêtres)")
  r.set_yticklabels(['', 'transition','assis','escaliers','babyfoot','vélo','conduite','déjeuner','marche', 'travail'])
  r.set(xticks=([0, 500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500]))
```
